```python
import os
import pandas as pd

ministeries = ['fake', 'LNV', 'AZ', 'BuZa', 'BZK', 'EZK', 'FIN', 'IenW', 'JenV', 'OCW', 'SZW', 'VWS']
dfs = []

for number, file in enumerate(os.listdir()):
    if file.count('xlsx'):
        
        df = pd.read_excel(file)
        df['Ministerie'] = ministeries[number]
        df.columns = [column.strip() for column in df.columns]
        dfs.append(df)

df = pd.concat(dfs).drop('Kolom1', axis=1)
```

```python
df.to_csv('ondraaglijktraag.csv')
```
