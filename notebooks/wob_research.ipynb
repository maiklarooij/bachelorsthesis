```python
import pandas as pd

df = pd.read_csv('../data/wobs_rijksoverheid.csv', index_col=0)
df['pdfs'] = df['pdfs'].map(lambda x: x.strip("'][").split(', '))
df['number_of_documents'] = df['pdfs'].map(len)
df
```

```python
verantwoordelijk = df.groupby(['verantwoordelijk']).count()
verantwoordelijk['procent intro'] = verantwoordelijk['intro'] / verantwoordelijk['link'] * 100

verantwoordelijk
```

```python
verantwoordelijk.link.plot(kind="pie", title="Aantal behandelde Wob-verzoeken per ministerie", autopct='%1.0f%%')
```

```python
def get_doc_type(doclink):

    prefix = 'https://www.rijksoverheid.nl/binaries/rijksoverheid/documenten/wob-verzoeken'
    doclink = doclink.lower().strip(prefix)
    
    options = {'Inventaris': ['inventaris'], 'Bijlage': ['bijlage', 'document', 'agenda', 'verslag'], 'Besluit': ['besluit', 'beslissing']}
    
    for option in options:
        for keyword in options[option]:
            if keyword in doclink:
                return option

    return 'Niet herkend'
```

```python
df['Besluit'] = 0
df['Bijlage'] = 0
df['Inventaris'] = 0
df['Niet herkend'] = 0

for index, wob in df['pdfs'].items():
    for pdf in wob:
        doctype = get_doc_type(pdf)

        df[doctype][index] += 1
```

```python
for type_doc in ['Besluit', 'Bijlage', 'Inventaris', 'Niet herkend']:
    type_df = df[df[type_doc] != 0]
    verantwoordelijk[f"{type_doc} aantal"] = type_df.groupby(['verantwoordelijk']).count()[type_doc]
    verantwoordelijk[f"{type_doc} percentage"] = verantwoordelijk[f"{type_doc} aantal"] / verantwoordelijk['link'] * 100

verantwoordelijk.fillna(0)
```

```python
verantwoordelijk[['Besluit percentage', 'Bijlage percentage', 'Inventaris percentage']].sort_values(by=['Besluit percentage'], ascending=False).plot(kind='bar',
    title="Percentage van Wob-verzoeken per type document gegroepeerd op ministerie")
```

```python
import re
import pdfplumber
from collections import Counter
from tqdm.notebook import tqdm
import zipfile
import os

def clean(input_string):
    """
    Removes unneccesary characters (non-alphabetical and -numerical) from a string.
    """
    if type(input_string) != str:
        return None
    return re.sub('[\W_]+', '', input_string).lower()


def classify_column(column_name):
    """
    Tries to generalize some common column names.
    """
    if not column_name:
        return None
                
    options = {
        "nr": ["nr", "nummer", "volgnummer", "docnr", "documentnr", "id"],
        "naamdocument": ["document", "documentnaam", "titeldocument", "titeldoc", "onderwerp", "naamdocument", "titel", "naam"],
        "beoordeling": ["beoordeling", "beroordeling", "oordeel", "beoordelingwob"],
        "weigeringsgrond": ["weigeringsgrond", "artikelwob", "wob", "beslissingconform", "wobgrond", "uitzonderingsgrond", "artikel", "wobartikel", "weigeringsgrondwob", "weigeringsgronden",
                            "lakgrond", "relevantewobgronden", "grond"],
        "datum": ["datum", "datumdocument"],
        "soort": ["soort", "soortdocument", "type", "categorie", "documenttype", "typedocument", "soortstuk"],
        "afzender": ["afzender", "afzenders", "van"],
        "ontvanger": ["ontvanger", "ontvangers", "naar"]
    }

    for option in options:
        if (column_name in options[option]) or (column_name[::-1] in options[option]):
            return option
    
    return column_name

def get_column_row_from_table(table):

    result = table[0]

    how_many_cols_filled = len([col for col in table[0] if col])

    if how_many_cols_filled < int(len(table[0]) / 2):
        try:
            result = get_column_row_from_table(table[1:])
        except:
            return []

    return result


def retrieve_columns(inventaris_folder):
    """
    Tries to extract all column names from a zip file with inventarissen.
    """

    kolommen = []
    aantal = {}
    not_found = 0
    for file in tqdm(os.listdir(inventaris_folder)):
        
        table_found = False

        # Try to open the pdf file
        try:
            pdf = pdfplumber.open(inventaris_folder + file)
        except: 
            continue

        # Searches the first row of the first table found in pdf
        for page in pdf.pages:
            table=page.extract_table()

            if table:
                row = get_column_row_from_table(table)
                for kolom in row:
                    kolommen.append(classify_column(clean(kolom)))
                
                table_found = True
                break
        
        if not table_found:
            not_found += 1
            
    print(f"{not_found} documenten waarin geen tabel is gevonden.")

    return Counter(kolommen).most_common()
```
