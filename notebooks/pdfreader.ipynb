## pdfplumber

- Seems to read in PDF tables quite well
- Has some difficulties with reading PDF's when there is not only a table

```python
pdf = pdfplumber.open("data/5d03dc8b8621ea5e53ad6fe7b44c8533_bijlage-4-inventarislijst-deelbesluit-1.pdf")

df = pd.DataFrame()
for page in pdf.pages:
    table=page.extract_table()
    df = df.append(pd.DataFrame(table[1::],columns=pdf.pages[0].extract_table()[0]))
df
```
